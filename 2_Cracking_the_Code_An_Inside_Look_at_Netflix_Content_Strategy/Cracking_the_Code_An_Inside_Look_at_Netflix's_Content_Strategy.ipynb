### Step 5: Feature Engineering - Content Freshness

#### 2. Is there a relationship between content age and its type (Movie vs. TV Show)?

```python
# We already calculated 'age_on_netflix'. Now we'll compare the distribution of this feature for Movies and TV Shows.

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create the 'age_on_netflix' feature if it doesn't exist
if 'age_on_netflix' not in df.columns:
    df['age_on_netflix'] = df['year_added'] - df['release_year']

# Filter out any potential errors where added_year is before release_year
content_age = df[df['age_on_netflix'] >= 0]

plt.figure(figsize=(14, 7))
sns.boxplot(data=content_age, x='type', y='age_on_netflix', palette='coolwarm', hue='type', legend=False)
plt.title('Content Age Distribution by Type (Movie vs. TV Show)')
plt.xlabel('Content Type')
plt.ylabel('Content Age (Years)')
plt.show()
```

#### 4. What are the most common word pairs or phrases in content descriptions?

```python
# To find common word pairs (bigrams), we can use libraries like NLTK or implement it manually.
# First, we'll clean the text by removing punctuation and converting to lowercase.
# Then, we'll generate bigrams and count their frequency.

import nltk
from nltk.corpus import stopwords
from collections import Counter
import string
import pandas as pd

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab') # Download the missing resource

# Combine all descriptions
text = ' '.join(df['description']).lower()

# Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Tokenize the text into words
words = nltk.word_tokenize(text)

# Remove stop words
stop_words = set(stopwords.words('english'))
words = [word for word in words if word not in stop_words]

# Generate bigrams (word pairs)
bigrams = list(nltk.bigrams(words))

# Count the frequency of bigrams
bigram_counts = Counter(bigrams)

# Get the most common bigrams
top_bigrams = bigram_counts.most_common(20)

# Print the top bigrams
print("Top 20 most common word pairs (bigrams):")
for bigram, count in top_bigrams:
    print(f"{bigram[0]} {bigram[1]}: {count}")
```

**Insights Answer 4:**

Based on the output of the most common word pairs (bigrams), the content descriptions frequently feature phrases like **"high school," "young man," "young woman," "new york," "best friends," and "true story**." These bigrams suggest that common themes in Netflix content include **coming-of-age stories, narratives set in urban environments (specifically New York), stories centered around friendships, and content based on real events. Other common pairs like "standup special," "world war," "documentary follows," and "serial killer"** highlight specific popular genres and subjects available on the platform.
